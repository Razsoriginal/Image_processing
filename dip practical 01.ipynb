```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import img_as_uint
from skimage.io import imread
from skimage.color import rgb2hsv, rgb2gray
```

```python
image_path = r"C:/Users/shiva/Downloads/dog.jpg"
doggo = imread(image_path)
```

```python
# Display the original RGB image
plt.imshow(doggo)

# Remove axis values for better visualization
plt.axis('off')

# Show the image
plt.show()
```

```python
# Display the shape of the image
# Output format: (height, width, color_channels)
doggo.shape
```

```python
# Create a figure with 3 subplots to display image slices
fig, ax = plt.subplots(1, 3, figsize=(6, 4), sharey=True)

# First vertical slice of the image
ax[0].imshow(doggo[:, 0:130])
ax[0].set_title('First Split')

# Second vertical slice of the image
ax[1].imshow(doggo[:, 130:260])
ax[1].set_title('Second Split')

# Third vertical slice of the image
ax[2].imshow(doggo[:, 260:390])
ax[2].set_title('Third Split')

# Display all slices
plt.show()
```

```python
# Crop a specific region from the image
# Format: image[row_start:row_end, col_start:col_end]
plt.imshow(doggo[95:250, 130:275])

# Hide axis for clarity
plt.axis('off')

# Show cropped image
plt.show()
```

```python
# Create subplots to visualize Red, Green, and Blue channels
fig, ax = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

# Display Red channel
ax[0].imshow(doggo[:, :, 0], cmap='Reds')
ax[0].set_title('Red Channel')

# Display Green channel
ax[1].imshow(doggo[:, :, 1], cmap='Greens')
ax[1].set_title('Green Channel')

# Display Blue channel
ax[2].imshow(doggo[:, :, 2], cmap='Blues')
ax[2].set_title('Blue Channel')

# Show RGB components
plt.show()
```

```python
# Convert the RGB image into HSV color space
# HSV separates color information (Hue) from intensity (Value)
doggo_hsv = rgb2hsv(doggo)
```

```python
# Display Hue, Saturation, and Value components separately
fig, ax = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

# Hue component
ax[0].imshow(doggo_hsv[:, :, 0], cmap='hsv')
ax[0].set_title('Hue')

# Saturation component
ax[1].imshow(doggo_hsv[:, :, 1], cmap='gray')
ax[1].set_title('Saturation')

# Value component
ax[2].imshow(doggo_hsv[:, :, 2], cmap='gray')
ax[2].set_title('Value')

# Show HSV components
plt.show()
```

```python
# Convert the RGB image into grayscale
# Grayscale image contains intensity information only
doggo_gray = rgb2gray(doggo)
```

```python
# Display the grayscale image and its thresholded versions
fig, ax = plt.subplots(1, 5, figsize=(17, 6), sharey=True)

# Original grayscale image
ax[0].imshow(doggo_gray, cmap='gray')
ax[0].set_title('Grayscale Original')

# Thresholding using different intensity values
ax[1].imshow(img_as_uint(doggo_gray > 0.25), cmap='gray')
ax[1].set_title('Intensity > 0.25')

ax[2].imshow(img_as_uint(doggo_gray > 0.50), cmap='gray')
ax[2].set_title('Intensity > 0.50')

ax[3].imshow(img_as_uint(doggo_gray > 0.75), cmap='gray')
ax[3].set_title('Intensity > 0.75')

# Thresholding using mean intensity
ax[4].imshow(img_as_uint(doggo_gray > np.mean(doggo_gray)), cmap='gray')
ax[4].set_title('Intensity > Mean')

# Display thresholded images
plt.show()
```

```python
# Create a random grayscale image of size 256x256
image_path = r"C:/Users/shiva/Downloads/lina.jpeg"
img = rgb2gray(imread(image_path))
#plt.imshow(img)
#doggo_gray = rgb2gray(img)
#img = np.random.rand(256, 256)

# Reduce quantization to 2 bits (4 gray levels)
reduced_img_2bits = np.round(img * 3) / 3

# Reduce quantization to 1 bit (2 gray levels)
reduced_img_1bit = np.round(img)

# Display original and quantized images
plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(reduced_img_2bits, cmap='gray')
plt.title('2 bits (4 levels)')

plt.subplot(1, 3, 3)
plt.imshow(reduced_img_1bit, cmap='gray')
plt.title('1 bit (2 levels)')

plt.show()
```

```python
# Create a random grayscale image
image_path = r"C:/Users/shiva/Downloads/lina.jpeg"
img = rgb2gray(imread(image_path))
#img = np.random.rand(256, 256)

# Reduce sampling rate to half by skipping every alternate pixel
reduced_img_half = img[::2, ::2]

# Reduce sampling rate to quarter by skipping more pixels
reduced_img_quarter = img[::4, ::4]

# Display original and sampled images
plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(reduced_img_half, cmap='gray')
plt.title('Half Resolution')

plt.subplot(1, 3, 3)
plt.imshow(reduced_img_quarter, cmap='gray')
plt.title('Quarter Resolution')

plt.show()
```

```python
# Create a random grayscale image
image_path = r"C:/Users/shiva/Downloads/lina.jpeg"
img = rgb2gray(imread(image_path))
#img = np.random.rand(256, 256)

# Apply both sampling and quantization (2 bits, half resolution)
reduced_img_2bits_half = np.round(img[::2, ::2] * 3) / 3

# Apply both sampling and quantization (1 bit, quarter resolution)
reduced_img_1bit_quarter = np.round(img[::4, ::4])

# Display original and processed images
plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(reduced_img_2bits_half, cmap='gray')
plt.title('2 bits, Half Resolution')

plt.subplot(1, 3, 3)
plt.imshow(reduced_img_1bit_quarter, cmap='gray')
plt.title('1 bit, Quarter Resolution')

plt.show()
```
