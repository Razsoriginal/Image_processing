# Practical 2A (Arithmetics operations on image)

```python
import cv2
import numpy as np
from IPython import get_ipython
from IPython.display import display
import matplotlib.pyplot as plt
from skimage import data

coins_image = data.coins()
coins_image_cv2 = cv2.cvtColor(coins_image, cv2.COLOR_GRAY2BGR)

astronaut_image= data.astronaut()
astranaut_image_cv2 = cv2.cvtColor(astronaut_image,cv2.COLOR_RGB2BGR)

cv2.imwrite("coins_image.jpg",coins_image_cv2)
cv2.imwrite("astranaut_image.jpg",astranaut_image_cv2)

image1 = cv2.imread("coins_image.jpg")
image2 = cv2.imread("astranaut_image.jpg")

print(f"Original image 1 shape {image1.shape}")
print(f"Original image 2 shape {image2.shape}")

image_1_resized = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

added_image = cv2.add(image_1_resized, image2)
subtracted_image = cv2.subtract(image_1_resized, image2)
multiplied_image = cv2.multiply(image_1_resized, image2)
divided_image = cv2.divide(image_1_resized, image2)

fig ,axes = plt.subplots(3,3,figsize = (15,15))

axes[0,0].imshow(cv2.cvtColor(image1,cv2.COLOR_BGR2RGB))
axes[0,0].set_title(f"Original Image 1 {image1.shape[1]} {image1.shape[0]}")
axes[0,0].axis('off')

axes[0,1].imshow(cv2.cvtColor(image2,cv2.COLOR_BGR2RGB))
axes[0,1].set_title(f"Original Image 2 {image2.shape[1]} {image2.shape[0]}")
axes[0,1].axis('off')


axes[0,2].imshow(cv2.cvtColor(image_1_resized, cv2.COLOR_BGR2RGB))
axes[0,2].set_title(f"Resized Image 1 {image_1_resized.shape[1]}x {image_1_resized.shape[0]}")
axes[0,2].axis('off')

axes[1,0].imshow(cv2.cvtColor(added_image,cv2.COLOR_BGR2RGB))
axes[1,0].set_title(f"Added image")
axes[1,0].axis('off')

axes[1,1].imshow(cv2.cvtColor(subtracted_image,cv2.COLOR_BGR2RGB))
axes[1,1].set_title("Subtracted Image")
axes[1,1].axis('off')

axes[1,2].axis('off')

axes[2,0].imshow(cv2.cvtColor(multiplied_image,cv2.COLOR_BGR2RGB))
axes[2,0].set_title("Multiplied image")
axes[2,0].axis('off')

axes[2,1].imshow(cv2.cvtColor(divided_image,cv2.COLOR_BGR2RGB))
axes[2,1].set_title("Divided image")
axes[2,1].axis('off')

axes[2,2].axis('off')

plt.tight_layout()
plt.show()
```

# Practical 2B (Discrete Cosine Transform)

```python
import cv2
import numpy as np
from IPython import get_ipython
from IPython.display import display
import matplotlib.pyplot as plt
from skimage import data

coins_image = data.coins()
coins_image_cv2 = coins_image

dct_image = cv2.dct(np.float32(coins_image_cv2))
idct_image = cv2.idct(dct_image)


plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow((coins_image_cv2),cmap = "gray")
plt.axis('off')

plt.subplot(1,3,2)
plt.title("DCT Image")
plt.imshow(np.log(abs(dct_image) + 1), cmap="gray" )
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Reconstructed Image")
plt.imshow(idct_image, cmap = "gray")
plt.axis('off')

plt.show()
```
